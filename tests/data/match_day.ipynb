```python
import json
import pandas as pd

df = pd.read_json("match_day.json")
df.head()
df.columns.to_list()
```

```python
cols_cards = [
       "home_name", 
    "away_name",
    "team_a_yellow_cards", 
    "team_b_yellow_cards",
    "team_a_red_cards", 
    "team_b_red_cards",
    "team_a_cards_num", 
    "team_b_cards_num",
    "team_a_fh_cards",   # tarjetas 1er tiempo (si está registrado)
    "team_b_fh_cards",
    "team_a_2h_cards",   # tarjetas 2do tiempo
    "team_b_2h_cards",
    "total_fh_cards", 
    "total_2h_cards",
    "team_a_cards_0_10_min", 
    "team_b_cards_0_10_min",
    "cards_potential"    # probabilidad estimada de tarjetas
]
```

```python
# Cargar el archivo
with open("match_day.json", "r", encoding="utf-8") as f:
    data = json.load(f)

# Extraer la sección "data"
match_data = data["data"]

# Crear un DataFrame con solo las columnas relevantes para tarjetas
df_cards = pd.DataFrame([{col: match_data.get(col) for col in cols_cards}])

print(df_cards)
```
